```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## A simple example

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Plot some data on the axes.
_ = ax.plot([1,2,3,4], [3,1,5,1])
_ = ax.plot([1,2,3,4], [2,2,2,2])
```

**NOTE**: You cannot `fig.show()` and if `ax` is using the `plot` function in another cell it doesn't display it

```python
ax.plot([1,2,3,4], [2,5,1,3])
fig.show()
```
